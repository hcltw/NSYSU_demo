```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot
url = 'https://raw.githubusercontent.com/hcltw/NSYSU_demo/refs/heads/main/data/'
```

```python
#讀取15位受測者的資料
big_array = []

for i in range(1, 16):              # 受試者編號從 x 從 1 到 15
    filename = f"{i}.csv"
    filepath = os.path.join(url, filename)
    df = pd.read_csv(filepath, header=None)  
    arr = df.values               # 轉成 numpy array
    big_array.append(arr)
```
