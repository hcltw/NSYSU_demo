```python
import pandas as pd
import numpy as np
import os
url = 'https://raw.githubusercontent.com/hcltw/NSYSU_demo/refs/heads/main/data/'
```

```python
arrays = []

for i in range(1, 16):              # 受試者編號從 x 從 1 到 15
    filename = f"{i}.csv"
    filepath = os.path.join(url, filename)
    df = pd.read_csv(filepath, header=None)  # 第一列是資料
    arr = df.values               # 轉成 numpy array
    arrays.append(arr)
```

```python
#印出有幾個受試者
print(len(arrays))
```

```python
#印出1號受試者的第一行數據
print(arrays[0][1])
```
